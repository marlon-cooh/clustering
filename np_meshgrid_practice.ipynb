```
import numpy as np
import matplotlib.pyplot as plt
```

```
A = np.array([1,0])
B = np.array([0,1])
```

```
X, Y = np.meshgrid(A, B)
plt.figure(figsize=(4,4))
plt.scatter(X, Y)
plt.show()
```

```
print(X.ravel())
print(Y.ravel())
```

```
plt.figure(figsize=(4,4))
plt.scatter(X.ravel(), Y.ravel())
plt.grid(True)
plt.title('Grid Points Generated by np.meshgrid() and Flattened with .ravel()')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
